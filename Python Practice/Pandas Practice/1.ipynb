```python
# Sort desc according to index
newdf.sort_index(axis =0, ascending=False)
# axis = 0  indicate the rows
# By default ascending is kept as true
```

```python
newdf[0]
```

```python
# Creating a view of the dataframe
newdf2 = newdf
# Here newdf2 is not the copy of the newdf but a view of newdf2
# view can be considered like a pointer to the original variable
# Thus if any change is made to the newdf2 then it will also be reflected in the newdf dataframe
```

```python
newdf2[0][0] =1010
newdf
```

```python
# To make a copyu of newdy dataframe
newdf3 = newdf.copy()
# a change made in the copy will not be reflected in the original dataframe
```

```python
# Another way of updating the value at a particular index using loc()
newdf.loc[0,0] = 654

# This will update the value at 0th column of 0th row
# If index is not found then it will create a new row or column whichever index is not found and will initialize the row or column with value NaN        
```

```python
newdf
```

```python
# To delete a row 
newdf3.drop(0)
# This will drop row with index 0
```

```python
# To delete column we have to give extra parameter axis=1 
# axis = 1 represents column
# by default the axis is set to 0 which represents row
newdf3.drop(4,axis=1)
```

```python
# to get a particular set of rows and columns
newdf.loc[[0,1],[2,3]]
# This will output rows 0 and 1 with columns 2 and 3
```

```python
# Selecting rows where value less than 0.3
newdf3.loc[(newdf3[0])<0.3]
```

```python
# Multiple conditions
newdf3.loc[(newdf3[0]<0.3) & (newdf3[1]>0.5)]
```

```python
newdf3
```

```python
newdf3.iloc[0,4]
# In loc function name of the column can also be given
# But in iloc it is used with integer based searching
```
