```python
# Plot a single chart showing three lines one each for c-marks(red line), java-marks(blue marks) and py-marks(green line) against student name in python

import matplotlib.pyplot as plt

# Sample data
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
c_marks = [85, 78, 92, 88, 76]
java_marks = [79, 85, 88, 90, 82]
py_marks = [92, 89, 85, 87, 90]

# Plotting the data
# plt.figure(figsize=(10, 5))

plt.plot(students, c_marks, 'r-', label='C Marks')     # Red line
plt.plot(students, java_marks, 'b-', label='Java Marks') # Blue line
plt.plot(students, py_marks, 'g-', label='Python Marks') # Green line

# Adding titles and labels
plt.title('Student Marks in Different Subjects')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
name = ['A', 'B', 'C', 'D']
height = [150, 180, 140, 160]
weight = [55, 60, 40, 45]

# Define the bar width and the position of bars on the x-axis
bar_width = 0.35
index = np.arange(len(name))

# Create the bar plots
plt.bar(index, height, bar_width, label='Height', color='b')
plt.bar(index + bar_width, weight, bar_width, label='Weight', color='g')

# Adding titles and labels
plt.title('Height and Weight of Persons')
plt.xlabel('Person Name')
plt.ylabel('Value')
plt.xticks(index + bar_width / 2, name)

# Adding legend
plt.legend()

# Display the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample DataFrame
data = {
    'rollno': [1, 2, 3, 4, 5],
    'student_name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'cmarks': [85, 78, 92, 88, 76],
    'cppmarks': [80, 85, 88, 90, 79],
    'pymarks': [90, 88, 85, 91, 82]
}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
bar_width = 0.25
index = np.arange(len(df['student_name']))

plt.bar(index - bar_width, df['cmarks'], bar_width, label='C Marks', color='red')
plt.bar(index, df['cppmarks'], bar_width, label='C++ Marks', color='green')
plt.bar(index + bar_width, df['pymarks'], bar_width, label='Python Marks', color='blue')

plt.title('Marks of Students')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.xticks(index, df['student_name'])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot pie chart: Subject viz. count-of-students-who-scored-more-than-80-percentage
subjects = ['C', 'C++', 'Python']
counts = [len(df[df['c_marks'] > 80]), len(df[df['java_marks'] > 80]), len(df[df['py_marks'] > 80])]
plt.pie(counts, labels=subjects, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Subjects with Scores > 80%')
plt.show()
```
